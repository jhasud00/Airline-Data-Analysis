### Data Cleaning

```python
# Remove canceled flights
flights = flights[flights['CANCELLED'] == 0]

#Considering only round trips 
tickets = tickets[tickets['ROUNDTRIP'] == 1]

# Filter for medium and large airports
airport_codes = airport_codes[airport_codes['TYPE'].isin(['medium_airport', 'large_airport'])]
```

### Merging Airport Code and flight data set

```python
airport_codes = airport_codes.rename(columns={'IATA_CODE': 'AIRPORT_CODE', 'TYPE': 'AIRPORT_SIZE'})
flights = flights.merge(airport_codes, left_on='ORIGIN', right_on='AIRPORT_CODE', how='left', suffixes=('', '_ORIGIN'))
flights = flights.merge(airport_codes, left_on='ORIGIN', right_on='AIRPORT_CODE', how='left', suffixes=('', '_ORIGIN'))
flights = flights.merge(airport_codes, left_on='DESTINATION', right_on='AIRPORT_CODE', how='left', suffixes=('', '_DESTINATION'))
flights.columns
```

### Droping unnecessary column 

```python
columns_to_drop = [
     'AIRPORT_CODE', 'NAME', 'CONTINENT', 'ISO_COUNTRY', 'MUNICIPALITY', 'COORDINATES',
     'AIRPORT_CODE_ORIGIN', 'NAME_ORIGIN', 'CONTINENT_ORIGIN', 'ISO_COUNTRY_ORIGIN', 
     'MUNICIPALITY_ORIGIN', 'ELEVATION_FT_DESTINATION','COORDINATES_ORIGIN', 'AIRPORT_CODE_DESTINATION', 'NAME_DESTINATION', 
     'CONTINENT_DESTINATION', 'ISO_COUNTRY_DESTINATION', 'MUNICIPALITY_DESTINATION', 'COORDINATES_DESTINATION','ELEVATION_FT_ORIGIN',
]
# Retain only columns that exist in the dataframe
columns_to_drop = [col for col in columns_to_drop if col in flights.columns]

flights = flights.drop(columns=columns_to_drop)
```

### Changing data type of distance in flight data set

```python
# Inspect the 'DISTANCE' column for non-numeric values
non_numeric_distances = flights[~flights['DISTANCE'].astype(str).str.replace('.', '', 1).str.isdigit()]
print(non_numeric_distances)

# Replace non-numeric values with NaN and convert the column to float
flights['DISTANCE'] = pd.to_numeric(flights['DISTANCE'], errors='coerce')

# Drop rows with NaN values in 'DISTANCE' if necessary
flights.dropna(subset=['DISTANCE'], inplace=True)

# Now convert the column to float
flights['DISTANCE'] = flights['DISTANCE'].astype(float)
```

### Function of calculating total operational cost of flight

```python
def calculate_total_cost(row):
    distance = row['DISTANCE']
    
    # Fuel, Oil, Maintenance, Crew
    fuel_maintenance_crew_cost = 8 * distance
    
    # Depreciation, Insurance, Other
    depreciation_insurance_other_cost = 1.18 * distance
    
    # Airport Operational Costs
    airport_cost_origin = 5000 if row['AIRPORT_SIZE_ORIGIN'] == 'medium_airport' else 10000
    airport_cost_destination = 5000 if row['AIRPORT_SIZE_DESTINATION'] == 'medium_airport' else 10000
    
    # Delay Costs
    departure_delay_cost = max(0, row['DEP_DELAY'] - 15) * 75
    arrival_delay_cost = max(0, row['ARR_DELAY'] - 15) * 75
    
    total_cost = (fuel_maintenance_crew_cost + 
                  depreciation_insurance_other_cost + 
                  airport_cost_origin + 
                  airport_cost_destination + 
                  departure_delay_cost + 
                  arrival_delay_cost)
    
    return total_cost

# # Calculate total cost for each flight
# flights['calculated_total_cost'] = flights.apply(calculate_total_cost, axis=1)

# # Calculate average total cost grouped by origin and destination
# flights_calculated_df = flights.groupby(['ORIGIN', 'DESTINATION'])['calculated_total_cost'].mean().reset_index(),['OCCUPANCY_RATE'].mean().reset_index()

# # Rename columns for clarity
# avg_calculated_cost_df.columns = ['ORIGIN', 'DESTINATION', 'Average Total Cost']
# avg_calculated_cost_df.head()
```

### Calculating the all the calulated columns from flight dataframe

# Calculate total cost for each flight
flights['calculated_total_cost'] = flights.apply(calculate_total_cost, axis=1)

# Calculate average total cost , avg occupancy rate , avg arrival delay grouped by origin and destination
flights_calculated_df = flights.groupby(['ORIGIN', 'DESTINATION'])['calculated_total_cost'].mean().reset_index()
avg_occupancy_rate_df = flights.groupby(['ORIGIN', 'DESTINATION'])['OCCUPANCY_RATE'].mean().reset_index()
avg_delay_df = flights.groupby(['ORIGIN', 'DESTINATION'])['ARR_DELAY'].mean().reset_index()
total_trip_df = flights.groupby(['ORIGIN', 'DESTINATION']).size().reset_index(name='count')

flight_profit_data = pd.merge(flights_calculated_df, avg_occupancy_rate_df, left_on=['ORIGIN', 'DESTINATION'], right_on=['ORIGIN', 'DESTINATION'], how='inner')
flight_profit_data = pd.merge(flight_profit_data, avg_delay_df, left_on=['ORIGIN', 'DESTINATION'], right_on=['ORIGIN', 'DESTINATION'], how='inner')
flight_profit_data = pd.merge(flight_profit_data, total_trip_df, left_on=['ORIGIN', 'DESTINATION'], right_on=['ORIGIN', 'DESTINATION'], how='inner')
# Rename columns for clarity
flight_profit_data.columns = ['ORIGIN', 'DESTINATION', 'Average Oeperational Cost','Avg Occupancy Rate','Avg_arr_delay','Total_trip']
flight_profit_data.head()

### Calculating the all the calulated columns from ticket dataframe

```python
tickets['ITIN_FARE'] = pd.to_numeric(tickets['ITIN_FARE'], errors='coerce')

# Assuming the price listed is for a one-way trip, calculate the round trip revenue
tickets['ROUND_TRIP_REVENUE'] = tickets['ITIN_FARE']

# Group by origin and destination to calculate the total and average round trip revenue
round_trip_revenue = tickets.groupby(['ORIGIN', 'DESTINATION']).agg(
    TOTAL_ROUND_TRIP_REVENUE=('ROUND_TRIP_REVENUE', 'sum'),
    TOTAL_PASSENGERS=('PASSENGERS', 'sum')
).reset_index()

# Calculate the average round trip revenue per passenger
round_trip_revenue['AVERAGE_ROUND_TRIP_REVENUE_PER_PASSENGER'] = round_trip_revenue['TOTAL_ROUND_TRIP_REVENUE'] / round_trip_revenue['TOTAL_PASSENGERS']

# Rename columns for clarity
round_trip_revenue.columns = ['ORIGIN', 'DESTINATION', 'Total Round Trip Revenue', 'Total Passengers', 'Average Round Trip Revenue Per Passenger']

# Display the top 10 routes by total round trip revenue
top_routes_by_passenger = round_trip_revenue.drop(columns=['Total Round Trip Revenue','Total Passengers'])
top_routes_by_passenger
```

### Merging all the calculated columns in one data frame  

```python

# Create a DataFrame for round trips by merging the flights data on origin and destination
round_trips = pd.merge(
    flight_profit_data,
    flight_profit_data,
    left_on=['ORIGIN', 'DESTINATION'],
    right_on=['DESTINATION', 'ORIGIN'],
    suffixes=('_out', '_return')
)
# Combine the columns for routes
round_trips['ORIGIN'] = round_trips['ORIGIN_out']
round_trips['DESTINATION'] = round_trips['DESTINATION_out']

# Calculate the sum of average total costs
round_trips['Average Oeperational Cost'] = round_trips['Average Oeperational Cost_out'] + round_trips['Average Oeperational Cost_return']

# Calculate the mean of average occupancy rates
round_trips['Mean_Occupancy_Rate'] = (round_trips['Avg Occupancy Rate_out'] + round_trips['Avg Occupancy Rate_return']) / 2

round_trips['Avg_arr_delay'] = (round_trips['Avg_arr_delay_out'] + round_trips['Avg_arr_delay_return']) / 2
round_trips['Total_trip'] = round_trips['Total_trip'] = round_trips[['Total_trip_out', 'Total_trip_return']].min(axis = 1)
# Create the final DataFrame
final_roundtrip_df = round_trips[['ORIGIN', 'DESTINATION', 'Average Oeperational Cost', 'Mean_Occupancy_Rate','Avg_arr_delay','Total_trip']]
final_roundtrip_df
# Filter to ensure we only consider valid round trips (ignoring same flight pairs)
#round_trips = round_trips[round_trips['FLIGHT_ORIGIN_out'] != round_trips['FLIGHT_ID_return']]
#round_trips
```

### Calculating profit of each round trips and total revenue

```python

# Merge the data frames on ORIGIN and DESTINATION
merged_df = pd.merge(final_roundtrip_df,top_routes_by_passenger, on=['ORIGIN', 'DESTINATION'])
# Calculate profit
round_trip_revenue = merged_df['Average Round Trip Revenue Per Passenger'] * merged_df['Mean_Occupancy_Rate'] * 200

# Calculate the additional revenue from the occupancy rate
additional_revenue = merged_df['Mean_Occupancy_Rate'] * 100 * 70

# Sum the revenues
total_revenue = round_trip_revenue + additional_revenue

# Calculate the profit
merged_df['PROFIT'] = total_revenue - merged_df['Average Oeperational Cost']

# Round the 'PROFIT' column to 2 decimal places
merged_df['PROFIT'] = merged_df['PROFIT'].round(2)

# Display the dataframe
merged_df

```

### Normalizing Origin and Destination to get the routes

```python
# Normalize routes by sorting the origin and destination pairs
merged_df['Normalized_Route'] = merged_df.apply(lambda x: '-'.join(sorted([x['ORIGIN'], x['DESTINATION']])), axis=1)

# Group by normalized routes and calculate the mean for all other columns
aggregated_df = merged_df.groupby('Normalized_Route').agg({
    'Average Oeperational Cost': 'mean',
    'Mean_Occupancy_Rate': 'mean',
    'Avg_arr_delay': 'mean',
    'Total_trip': 'mean',
    'Average Round Trip Revenue Per Passenger': 'mean',
    'PROFIT': 'mean'
}).reset_index()

aggregated_df
```

### Imported to CSV to answr the anlysis Question

#### Split the normalized route back into origin and destination for clarity

```python
aggregated_df[['ORIGIN', 'DESTINATION']] = aggregated_df['Normalized_Route'].str.split('-', expand=True)
aggregated_df = aggregated_df[['ORIGIN', 'DESTINATION', 'Average Oeperational Cost', 'Mean_Occupancy_Rate', 'Avg_arr_delay','Total_trip', 'Average Round Trip Revenue Per Passenger', 'PROFIT']]
aggregated_df = aggregated_df.rename(columns={
    'Total_trip': 'Total_roundTrips_per_quarter',
    'Average Round Trip Revenue Per Passenger': 'Avg_ticket_cost_per_passenger',
    'PROFIT': 'PROFIT_per_round_trip'
})
aggregated_df
```
